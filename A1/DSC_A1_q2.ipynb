## Q2 a)

```python
def mean_variance(data):

  sum = 0
  n = len(data)

  for value in data:
    sum += value

  mean = sum/n

  sum_sq_diff = 0

  for value in data:
    sum_sq_diff += (value - mean)**2

  variance = sum_sq_diff/n

  return mean, variance
```

```python
print('mean and variance of data ', mean_variance(data))
```

## Q2 b)

```python
def get_s1_s2_s3(data):

  sum = 0
  n = len(data)

  for value in data:
    sum += value

  mean = sum/n

  sum_sq_diff = 0

  for value in data:
    sum_sq_diff += (value - mean)**2

  s1 = sum_sq_diff/(n+1)
  s2 = sum_sq_diff/n
  s3 = sum_sq_diff/(n-1)

  return s1, s2, s3
```

## Q2 c)

```python
avg_s1 = []
avg_s2 = []
avg_s3 = []
```

```python
s1 = 0
s2 = 0
s3 = 0

iteration = 1500
for i in range(1, iteration+1):

  random_samples = np.random.choice(data, size=50, replace=True)
  temp = get_s1_s2_s3(random_samples)

  s1 += temp[0]
  s2 += temp[1]
  s3 += temp[2]

  avg_s1.append(s1/i)
  avg_s2.append(s2/i)
  avg_s3.append(s3/i)
```

Bias in Sample Variance:
When using the sample mean x̄ instead of the population mean μ in the formula for variance, there is a tendency for the squared differences (xi - x̄)² to be smaller on average than the squared differences (xi - μ)². This results in an underestimate of the population variance.

Bessel's Correction:
To correct for this bias and make the sample variance an unbiased estimator of the population variance, Bessel's correction suggests dividing the sum of squared differences by (n-1), where n is the sample size. This correction factor accounts for the reduction in degrees of freedom when estimating the population mean from the sample mean.

By dividing by (n - 1) instead of N, the sample variance gives a more accurate estimate of the population variance based on the available sample data. This correction ensures that, on average, the sample variance is not systematically lower than the population variance due to the use of the sample mean.
