## Q1 b)

```python
df_encoded = pd.read_csv("/content/drive/MyDrive/Data Science/Auto_MPG_encoded.csv")
```

```python
def feature_mean_variance(feature):
  sum = 0
  n = len(feature)

  for value in feature:
    sum += value

  mean = sum/n

  sum_sq_diff = 0

  for value in feature:
    sum_sq_diff += (value - mean)**2

  varience = sum_sq_diff/n

  return mean, varience
```

```python
df_encoded.columns
```

```python
df_encoded = df_encoded.astype(float)
```

```python
mean_variance = {}
for col in df_encoded.columns:
  temp = feature_mean_variance(df_encoded[col])
  mean_variance[col] = temp
  print(f'mean and varience of {col} ',temp)
```

## Q1 c)

```python
df_normalized = df_encoded
```

```python
for col in df_normalized:
  for i in range(len(df_normalized[col])):
    val = df_normalized[col][i]

    mean = mean_variance[col][0]
    var = mean_variance[col][1]

    val_norm = (val - mean)/(var**0.5)

    df_normalized[col][i] = val_norm
```
